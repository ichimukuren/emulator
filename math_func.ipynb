```python
import math
import numpy as np

def f1( xy_obj, xy_rob, u,v):
       
        
    l0 =  xy_rob[0]

    x0,y0 = xy_obj
    x0,y0 = int(x0),l0 - int(y0)
    

    a=(l0-x0)*v
    b=v*y0
    c=u*y0
    sqr=math.sqrt(a**2+b**2)
    print(a**2,b**2,a**2+b**2)

    fi=math.asin(a/sqr)
    alfa=math.acos(c/sqr)+fi    
    t=y0/(v*math.sin(alfa))
    print(t)
    xobj=x0+u*t
    yobj=y0
    xrob=(l0+v*math.cos(alfa)*t)
    yrob=v*math.sin(alfa)*t
    y=(xrob-l0)*math.tan(alfa)
    
#     if xrob<400:
#         xrob = 400
    return (int(xrob),int(xy_obj[1]))

def f2( xy_obj, xy_rob,u,v):
    l0 =  xy_rob[0]
    
    
    x0,y0 = xy_obj
    

    a=(l0-x0)*v
    b=v*y0
    c=u*y0
    sqr=math.sqrt(a**2+b**2)
    print(a**2,b**2,a**2+b**2)
    fi=math.asin(a/sqr)
    alfa=math.acos(c/sqr)+fi
    
    x = y0/(math.tan(alfa))+l0
    
#     if x<400:
#         x = 400
    return (int(x),int(xy_obj[1]))
  
        
```

```python
dist = lambda a,b: np.linalg.norm(np.array(a) - np.array(b))

box = np.int0([100,100])
robot = np.int0([500,0])
print(f"Точка коробки: {box}")
print(f"Точка робота: {robot}")
print()
u_box   = 300
v_robot = 2000
print(f"Скорость коробки: {u_box}")
print(f"Скорость робота: {v_robot}")
print()

predict = f1(box,robot,u_box,v_robot)
print()
print(f"Точка столкновения: {predict}")
print()
s_box = dist(box,predict)
print(f"Путь пройденный коробкой: {s_box}")

s_robot = dist(robot,predict)
print(f"Путь пройденный роботом: {s_robot}")
print()
t_box   = s_box   / u_box
print(f"Время пройденное коробкой: {t_box}")

t_robot = s_robot / v_robot
print(f"Время пройденное роботом: {t_robot}")
print()

print(f"Робот должен прихеать и подождать: {t_box - t_robot} секунды")
print()

if t_box >= t_robot:
    print("Хорошо! робот успел придти на место, перед коробкой.")
else:
    print("Мы упустили коробку! Робот едет до места встречи слишком долго!")
```
